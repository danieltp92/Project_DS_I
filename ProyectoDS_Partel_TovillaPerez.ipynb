# The principal question we will try to solve is if the Fix Value when the mexican peso have a depretion, will be related to the price value in the american and mexican price stocks of the three principal banks in those two countries

```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
url_jpm = "https://raw.githubusercontent.com/danieltp92/Sources/main/JPM_Historical_Data.csv"
data_jpm = pd.read_csv(url_jpm)
jpm_df = data_jpm.copy()
jpm_df['Date_time'] = jpm_df['Date'].astype('datetime64[ns]')
columns_name = ["Date_time", "Stock_Price_JPM(USD)", "Open_Stock_Price_JPM(USD)", "High_Stock_Price_JPM(USD)", "Low_Stock_Price_JPM(USD)", "Vol_Stock_JPM", "Change_Stock_JPM_Perc"]
jpm_df = jpm_df.reindex(columns=columns_name)
```

```python
url_bac = "https://raw.githubusercontent.com/danieltp92/Sources/main/BAC_Historical_Data.csv"
data_bac = pd.read_csv(url_bac)
bac_df = data_bac.copy()
bac_df['Date_time'] = bac_df['Date'].astype('datetime64[ns]')
columns_name = ["Date_time", "Stock_Price_BAC(USD)", "Open_Stock_Price_BAC(USD)", "High_Stock_Price_BAC(USD)", "Low_Stock_Price_BAC(USD)", "Vol_Stock_BAC", "Change_Stock_BAC_Perc"]
bac_df = bac_df.reindex(columns=columns_name)
```

```python
url_wfc = "https://raw.githubusercontent.com/danieltp92/Sources/main/WFC_Historical_Data.csv"
data_wfc = pd.read_csv(url_wfc)
wfc_df = data_wfc.copy()
wfc_df['Date_time'] = wfc_df['Date'].astype('datetime64[ns]')
columns_name = ["Date_time", "Stock_Price_WFC(USD)", "Open_Stock_Price_WFC(USD)", "High_Stock_Price_WFC(USD)", "Low_Stock_Price_WFC(USD)", "Vol_Stock_WFC", "Change_Stock_WFC_Perc"]
wfc_df = wfc_df.reindex(columns=columns_name)
```

```python
url_bbva = "https://raw.githubusercontent.com/danieltp92/Sources/main/BBVA_Historical_Data.csv"
data_bbva = pd.read_csv(url_bbva)
bbva_df = data_bbva.copy()
bbva_df['Date_time'] = bbva_df['Date'].astype('datetime64[ns]')
columns_name = ["Date_time", "Stock_Price_BBVA(MXN)", "Open_Stock_Price_BBVA(MXN)", "High_Stock_Price_BBVA(MXN)", "Low_Stock_Price_BBVA(MXN)", "Vol_Stock_BBVA", "Change_Stock_BBVA_Perc"]
bbva_df = bbva_df.reindex(columns=columns_name)
```

```python
url_sant = "https://raw.githubusercontent.com/danieltp92/Sources/main/SantMx_Historical_Data.csv"
data_sant = pd.read_csv(url_sant)
sant_df = data_sant.copy()
sant_df['Date_time'] = sant_df['Date'].astype('datetime64[ns]')
columns_name = ["Date_time", "Stock_Price_SantMx(MXN)", "Open_Stock_Price_SantMx(MXN)", "High_Stock_Price_SantMx(MXN)", "Low_Stock_Price_SantMx(MXN)", "Vol_Stock_SantMx", "Change_Stock_SantMx_Perc"]
sant_df = sant_df.reindex(columns=columns_name)
```

```python
url_bano = "https://raw.githubusercontent.com/danieltp92/Sources/main/BanorteMx_Historical_Data.csv"
data_bano = pd.read_csv(url_bano)
bano_df = data_bano.copy()
bano_df['Date_time'] = bano_df['Date'].astype('datetime64[ns]')
columns_name = ["Date_time", "Stock_Price_BanorteMx(MXN)", "Open_Stock_Price_BanorteMx(MXN)", "High_Stock_Price_BanorteMx(MXN)", "Low_Stock_Price_BanorteMx(MXN)", "Vol_Stock_BanorteMx", "Change_Stock_BanorteMx_Perc"]
bano_df = bano_df.reindex(columns=columns_name)
```

```python
df_banks = pd.merge(jpm_df, bac_df, how='outer', on='Date_time')
df_banks = pd.merge(df_banks, wfc_df, how='outer', on='Date_time')
df_banks = pd.merge(df_banks, bbva_df, how='outer', on='Date_time')
df_banks = pd.merge(df_banks, sant_df, how='outer', on='Date_time')
df_banks = pd.merge(df_banks, bano_df, how='outer', on='Date_time')
df_banks
```

# In the graph below the stocks are showing their behavior during the time.
# We can see a great variance between the stock of USA and Mexico.
# To see the correct behavior we need to convert the value of the stocks from mexican pesos to american dollars.

```python
plt.plot(df_banks['Date_time'], df_banks['Stock_Price_JPM(USD)'], label='J. P. Morgan')
plt.plot(df_banks['Date_time'], df_banks['Stock_Price_BAC(USD)'], label='Bank of America')
plt.plot(df_banks['Date_time'], df_banks['Stock_Price_WFC(USD)'], label='Wells Fargo')
plt.plot(df_banks['Date_time'], df_banks['Stock_Price_BBVA(MXN)'], label='BBVA', color = 'b')
plt.plot(df_banks['Date_time'], df_banks['Stock_Price_SantMx(MXN)'], label='Santander', color = 'r')
plt.plot(df_banks['Date_time'], df_banks['Stock_Price_BanorteMx(MXN)'], label='Banorte', color = 'y')
plt.legend(loc='upper left')
```

# The following code take two sources of the historical fix value that convert the daily value of an american dollar to mexican pesos.

```python
url_fix = "https://raw.githubusercontent.com/danieltp92/Sources/main/Usd_Mxn_Exch_Rate_historical.csv"
data_fix = pd.read_csv(url_fix)
fix_df = data_fix.copy()
fix_df['Date_time'] = fix_df['Date'].astype('datetime64[ns]')
columns_name = ["Date_time","Fix_Value(MXN)"]
fix_df = fix_df.reindex(columns=columns_name)
fix_df
```

```python
url_fix2 = "https://raw.githubusercontent.com/danieltp92/Sources/main/Usd_Mxn_Exch_Rate_historical_2.csv"
data_fix2 = pd.read_csv(url_fix2)
fix_df2 = data_fix2.copy()
fix_df2['Date_time'] = fix_df2['DATE'].astype('datetime64[ns]')
fix_df2['Fix_Value(MXN)'] = fix_df2['DEXMXUS']
fix_df2['Fix_Value(MXN)'] = fix_df2['Fix_Value(MXN)'].replace(['.'], 0)
fix_df2['Fix_Value(MXN)'] = fix_df2['Fix_Value(MXN)'].replace('%','').astype(np.float64)
columns_name = ["Date_time","Fix_Value(MXN)"]
fix_df2 = fix_df2.reindex(columns=columns_name)
fix_df2
```

```python
df_fix = pd.merge(fix_df, fix_df2, how='outer', on=['Date_time','Fix_Value(MXN)'])
df_fix
```

# Ploting the Fix Value information we can see that in some values we don't have data. This information need to be fixed.
# As we can see the value of a dollar increase during the years. Reflecting the depresiation of the mexican peso against the american dollar.

```python
plt.plot(df_fix['Date_time'], df_fix['Fix_Value(MXN)'], label='Fix Value(MXN)')
```

```python
df_banks_2 = pd.merge(df_banks, df_fix, how='left', on='Date_time')
df_banks_2
```

# The following code make the convertion of the mexican value stock to amercian dollars value.

```python
df_banks_2['Stock_Price_BanorteMx(USD)'] = df_banks_2['Stock_Price_BanorteMx(MXN)'] / df_banks_2['Fix_Value(MXN)']
df_banks_2['Stock_Price_SantMx(USD)'] = df_banks_2['Stock_Price_SantMx(MXN)'] / df_banks_2['Fix_Value(MXN)']
df_banks_2['Stock_Price_BBVA(USD)'] = df_banks_2['Stock_Price_BBVA(MXN)'] / df_banks_2['Fix_Value(MXN)']
df_banks_2
```

```python
df_banks_2['month'] = df_banks_2['Date_time'].dt.month_name()
df_banks_2['year'] = pd.DatetimeIndex(df_banks_2['Date_time']).year
cols_end = ['Date_time', 'month', 'year', 'Fix_Value(MXN)']
df_banks_3 = df_banks_2[[c for c in cols_end if c in df_banks_2] + [c for c in df_banks_2 if c not in cols_end]]
df_banks_3
```

# The following plot is a comparision of american and mexican stocks, all in american dollar values. All the mexican stocks are below of a 10 USD value.
# The american stocks price are bigger even when their value decrease during the years.

```python
plt.plot(df_banks_3['Date_time'], df_banks_3['Stock_Price_JPM(USD)'], label='J. P. Morgan')
plt.plot(df_banks_3['Date_time'], df_banks_3['Stock_Price_BAC(USD)'], label='Bank of America')
plt.plot(df_banks_3['Date_time'], df_banks_3['Stock_Price_WFC(USD)'], label='Wells Fargo')
plt.plot(df_banks_3['Date_time'], df_banks_3['Stock_Price_BBVA(USD)'], label='BBVA', color = 'b')
plt.plot(df_banks_3['Date_time'], df_banks_3['Stock_Price_SantMx(USD)'], label='Santander', color = 'r')
plt.plot(df_banks_3['Date_time'], df_banks_3['Stock_Price_BanorteMx(USD)'], label='Banorte', color = 'y')
plt.legend(loc='upper left')
```

```python
df_stocks = df_banks_3[['Date_time', 'month', 'year', 'Fix_Value(MXN)', 'Stock_Price_JPM(USD)', 'Stock_Price_BAC(USD)', 'Stock_Price_WFC(USD)', 'Stock_Price_BBVA(USD)', 'Stock_Price_SantMx(USD)', 'Stock_Price_BanorteMx(USD)']].copy()
df_stocks
```

```python
col_not = ['Date_time','month','year']
l=[x for x in df_stocks.columns if x not in col_not]
stocks = df_stocks[l]
stocks.head()
```

# The box plots below show the outliers and inliers of the values of the Fix Value and the stock values.
# We can see that the american stock values have less outliers than the mexican stock values.

```python
plt.figure(figsize=(15,14))
for i,j in zip(range(7),stocks.columns):
    plt.subplot(4,2,i+1)
    sns.boxplot(x=stocks[j])
    plt.title(j)
    plt.xlabel('')
```

# Leaving aside the outliers of the Fix Value, the next scatter plots show that the american stock values decrease their price when the Fix Value increases.
# We can see that when the Fix value is between 15 and 20, the stock price increase.
